```python
# list of softskill keywords
def countkeywords(keyword):
    num = [str(row).lower().count(keyword) for row in data['description']]
    return keyword,sum(num)
```

adding them to the dataframe

```python
softskills = ['communication skills', 
              'problem solving', 
              'attention to detail', 
              'interpersonal skills', 
              'decision making', 
              'teamwork']
for i in softskills:
    data[i] = data['description'].str.contains(i, regex=False)
```

```python
data
```

```python
years = ["2", "3", "4", "5", "10" ]
# list comprehension
experience = [f'{x} years' for x in years]
experience.append('1 year')
print(experience)
for years in experience:
    data[f'has_{years}'] = data['description'].str.contains(years, regex=False)
```

```python
degreesneed = ['bachelor', 'master', 'phd ', 'msc ', 'ba ','bsc ', 'b.s ', 'ph.d ', 'b.s.', 'b.a ']
for skill in degreesneed:
    data[f'has_{skill}'] = data['description'].str.contains(skill, regex=False)
```

```python
hard_skills = ['python', 'sql', 'power bi', 'excel', 'tableau', 'java', 'c++']
for skill in hard_skills:
    data[f'has_{skill}'] = data['description'].str.contains(skill, regex=False)
```

```python
# removing the zips from location
import re
# Replace digits in column with nothing
data['location'] = [re.sub("\d", "", str(x)) for x in data['location']]
# strip leading and trailing spaces
data['location'] = data['location'].str.strip()
```

```python
# adding city and state
data[['City', 'State']] = data['location'].str.split(',', expand=True)
```

```python
positions = ['machine learning engineer', 'data scientist', 'research analyst', 'data analyst', 'data engineer', 'research scientist' ]
for position in positions:
    data[position] = data['position'].str.contains(position, regex=False)
```

```python
data['description'] = data['description'].str.replace('.', '')
data['description'] = data['description'].str.replace(',', '')
data['description'] = data['description'].str.replace(';', '')
```
