# Numerical & Categoricl Columns

Currently "Income", "Monthly Premium Auto" and "Total Claim Amount" are numerical. All the other columns are currently categorical. "Customer Lifetime Value" should also be numerical and presumably "Number of Open Complaints"


# Data Cleaning

```python
compilation = compilation.drop(['Education', 'Number of Open Complaints'], axis=1)
compilation
```

```python
# Get rid of percentage sign
for x in compilation["Customer Lifetime Value"]:
    if type(x) == str:
        x1 = x.strip('%')
        compilation["Customer Lifetime Value"] = compilation["Customer Lifetime Value"].replace(x, x1)
        
#Reset index
compilation.reset_index(drop=True, inplace=True)
```

```python
compilation
```

```python
compilation["Customer Lifetime Value"] = compilation["Customer Lifetime Value"].astype("float")*100
```

```python
compilation
```

# Remove Duplicates

```python
# Drop duplicates (where the entire row is identical)
compilation = compilation.drop_duplicates()
```
