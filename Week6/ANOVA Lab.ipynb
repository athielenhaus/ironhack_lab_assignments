# ANOVA Lab

## Part 1

#### Null hypothesis

H0 : changing the power of the plasma beam has no effect on the etching rate by the machine

#### Alternate hypothesis

H1 : changing the power of the plasma beam has an effect on the etching rate by the machine

#### Significance Level

The significance level, also denoted as alpha or Î±, is the probability of rejecting the null hypothesis when it is true.

#### Degrees of freedom of model, error terms, and total DoF

In statistics, the number of degrees of freedom is the number of values in the final calculation of a statistic that are free to vary.


## Part 2

```python
import pandas as pd
data= pd.read_excel("/Users/arnet/Desktop/Ironhack/GitHub/ironhack_lab_assignments/Week6/anova_lab_data.xlsx")
data
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols # ordinary least squares
model = ols('Etching_Rate ~ C(Power)', data = data).fit()
```

```python
results_table = sm.stats.anova_lm(model)
results_table
```

Degrees of Freedom: 
Model: 2
Error: 12
Total: 14

Significance < 0.000008

We can reject the null hypothesis that changing the power level has no effect, i.e. changing the power level does have an effect on the etching rate.
