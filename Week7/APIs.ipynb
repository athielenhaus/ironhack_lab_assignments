```python
import requests
```

```python
google = requests.get('http://developers.google.com')
print('google: ', google.status_code)
```

```python
nba = requests.get('http://api.sportsdata.io/api/nba/fantasy/json/CurrentSeason')
print('nba: ', nba.status_code)
```

```python
rotten = requests.get('http://api.rottentomatoes.com/api/public/v1.0/lists/movies')
print('rotten:', rotten.status_code)
```

```python
git = requests.get('https://api.github.com/users/athielenhaus/events/public')
git.status_code
git.json()
```
